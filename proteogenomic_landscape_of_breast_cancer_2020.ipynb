# Multi-layered Immune Escape Analysis in Breast Cancer (CPTAC 2020)

1. Project Overview and Data Source
This study investigates the mechanisms by which breast cancer (BRCA) tumors evade immune surveillance. We focus on the "Immune Escape Convergence" theory—the idea that tumors utilize multiple, potentially redundant strategies to hide their neoantigens from T-cell recognition.

Data Source
The analysis is based on the CPTAC Breast Cancer (Cell 2020) dataset, accessed via cBioPortal. The cohort includes 122 patients with complete multi-omic profiling, allowing for the integration of:

Genomics (WES): Somatic mutation data to identify potential neoantigens.

Transcriptomics (RNA-Seq): Gene expression levels to assess mutation silencing and HLA regulation.

Copy Number Alterations (CNA): Genomic gain/loss data to identify physical deletions of the antigen presentation machinery.

2. # Analysis of Immune Escape Mechanisms
We analyzed the tumor's "Lines of Defense" categorized into three distinct layers:

## Line 1: Transcriptomic Silencing of Mutations (Stealth Mode)
The first mechanism of escape is the suppression of genes harboring somatic mutations. If a mutated gene is not expressed, the resulting neoantigen will never be produced, making the mutation "invisible" to the immune system.

Methodology: We calculated the Mutation Silencing Ratio for each patient, defined as the proportion of somatic mutations occurring in genes with low or zero expression (<1.0 FPKM).

Key Finding: Approximately 34.7% of identified mutations were transcriptomically silenced. This suggests a significant "Stealth" strategy where the tumor avoids presenting a large portion of its mutational load.

In the first phase of our investigation, we performed a comprehensive multi-omic integration of genomic somatic mutations (MAF) and transcriptomic expression profiles (FPKM) for the cohort of 122 BRCA patients. The goal was to quantify "Transcriptomic Silencing"—a stealth strategy where the tumor avoids immune detection by suppressing the expression of genes harboring mutations. Our analysis revealed a significant global silencing rate of 34.7%, indicating that more than one-third of the tumor's genetic alterations are effectively "invisible" to the immune system at the transcript level.

Further granular analysis of the mutated genes revealed a clear evolutionary trade-off within the tumor's survival strategy. We observed a high frequency of silencing in large "passenger" genes, such as TTN, MUC16, and OBSCN, which are not essential for tumor viability and can thus be suppressed to minimize the presentation of neoantigens. Conversely, critical "driver" genes like TP53, PIK3CA, and GATA3 showed consistently low silencing rates. Because these drivers are essential for oncogenic signaling and proliferation, the tumor is forced to maintain their expression even at the cost of exposing their mutations to the immune system. This "driver exposure" creates a critical vulnerability in the tumor's first line of defense, serving as the likely evolutionary pressure that necessitates the transition to the more radical secondary escape mechanisms, such as the B2M genomic deletions identified in our subsequent Line 3 analysis.

# Line 2

In the second stage of our investigation, we analyzed the transcriptional regulation of the MHC class I machinery—specifically the HLA-A, HLA-B, HLA-C, and B2M genes. Our objective was to determine if tumors failing to silence their mutations (the Low Silencing group) compensated for this exposure by actively downregulating the expression of the antigen-presentation complex. We compared the RNA-seq expression profiles of these core components between the High Silencing and Low Silencing groups to identify any significant transcriptional shifts.

The analysis demonstrated that the expression levels of these core HLA and B2M genes were nearly identical across both cohorts. We observed no significant correlation between mutational visibility and the transcriptional suppression of the MHC-I complex, as the expression remained stable regardless of the silencing ratio. Consequently, we achieved the conclusion that selective transcriptional downregulation was not a primary mechanism of immune escape in this cohort. This lack of RNA-level adaptation suggests that the tumors did not utilize "software-level" control to hide their exposed mutations, highlighting the necessity for the physical, genomic deletions identified in our subsequent Line 3 analysis.

# Line 3

In the final stage of our study, we transitioned from transcriptional regulation to a genomic "hardware" analysis by examining Copy Number Alterations (CNA) within the antigen presentation machinery. Our investigation focused on identifying physical deletions in the B2M gene and the HLA-A, -B, -C locus, which represent a permanent, irreversible method of immune escape. We achieved a significant finding: 37.7% of the cohort (46 out of 122 patients) exhibited genomic deletions of B2M, which serves as the essential light chain for all MHC-I molecules. This identifies B2M as a critical "kill-switch" for the tumor; by physically deleting this single, stable gene on Chromosome 15, the tumor effectively collapses the entire display system regardless of HLA expression levels. Furthermore, we observed that while HLA genes themselves were absent from the standard CNA reports—a common result of the high polymorphism on Chromosome 6 that often leads to their filtering in standard bioinformatics pipelines—the high prevalence of B2M deletions provides a robust alternative strategy. This confirms that for a substantial portion of the cohort, physical destruction of the MHC-I complex via B2M loss is the definitive mechanism for ensuring total immune invisibility.

🏁 Synthesis: The Multi-layered Architecture of Immune Escape
By integrating these three lines of analysis, we achieved a comprehensive view of the immune escape landscape within the analyzed cohort. Our findings demonstrate that tumors do not rely on a single mechanism but instead employ an additive, multi-layered strategy to ensure survival. While Line 1 revealed that the tumors successfully hide approximately 34.7% of their mutational burden through transcriptomic silencing, this "stealth" mode is inherently limited by the necessity of expressing mutated driver genes. Line 2 exposed a transcriptional stalemate, where the tumors were unable to adaptively downregulate the MHC-I machinery to compensate for these exposed mutations, maintaining stable HLA levels regardless of their silencing efficiency. This failure of regulatory adaptation appears to drive the tumor toward the radical "hardware" solution found in Line 3: the physical, genomic deletion of B2M, observed in 37.7% of patients. Crucially, our synthesis shows that these strategies are not mutually exclusive; the genomic deletion of B2M occurs frequently even in tumors that already possess high silencing ratios. This complex interplay suggests that the tumors in this cohort utilize a redundant defense architecture—hiding what they can via RNA suppression while physically dismantling the antigen-presentation machinery to neutralize the remaining visible threats, ultimately creating a highly robust phenotype of immune invisibility.
